```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f,axis = plt.subplots(figsize=(18,18))
sns.heatmap(dframe.corr(), annot=True, linewidths=4, fmt= '.2f', ax=axis)
plt.show()
```

```python
# From the graph above, we can infer that the lighter the color gets, the stronger a correlation is.
# Average price of an avocado does not correlate to number of bags
# As the total volume gets larger, the number of small bags increases faster than larger bags 
```

```python
dframe.plot(kind = 'scatter', x = 'Small Bags', y = 'Total Volume', color = 'magenta', alpha = '0.6')
plt.xlabel('Small Bags')
plt.ylabel('Total Volume')
plt.title("Comparison Between Total Volume and Small Bags")
plt.show()
```
